```python
from bs4 import BeautifulSoup, Comment
import requests
import pandas as pd
```

```python
webpage = requests.get('https://www.hockey-reference.com/leagues/NHL_2018.html')
```

```python
soup = BeautifulSoup(webpage.text, 'lxml')
```

```python
table_soup = soup.find('div', {'id':'all_stats'})
table_soup
```

```python
table_lists = []
for comment in table_soup.find_all(string=lambda text:isinstance(text,Comment)):
    data = BeautifulSoup(comment,"lxml")
    for items in data.select("table.stats_table tr"):
        tds = [item.get_text(strip=True) for item in items.select("th,td")]
        table_lists.append(tds)
table_lists[0]
```

```python
del table_lists[0]
```

```python
columns = table_lists[0]
```

```python
df = pd.DataFrame(table_lists[1:], columns=columns)
```

```python
df['Year'] = 2018
df
```
