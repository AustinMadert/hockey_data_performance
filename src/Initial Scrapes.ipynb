```python
import requests
from bs4 import BeautifulSoup
```

```python
webpage = requests.get('http://www.nhl.com/ice/draftsearch.htm?year=&team=&position=&round=1')
```

```python
url = 'http://www.nhl.com/ice/draftsearch.htm?year=&team=&position=&round=1'
```

```python
webpage
```

```python
nhl_soup = BeautifulSoup(webpage.text, 'html.parser')
```

```python
nhl_tbodies = nhl_soup.find_all('tbody')
```

```python
nhl_tds = [obj.find_all('td') for obj in nhl_tbodies]
nhl_tds[0]
```

```python
parsed_site = [{ind: ''.join(str(val).strip('[').strip(']'))} for ind, val in enumerate([obj.find_all('td') for obj in nhl_tbodies])]
parsed_site[0]
```

```python
type(parsed_site[0])
```

```python
espn_page = requests.get('http://www.espn.com/nhl/team/schedule/_/name/wsh/season/2018/seasontype/2')
```

```python
espn_table = espn_soup.find('tr', class_='Table2__tr.Table2__tr--sm.Table2__even')
type(espn_table)
```

```python
espn_tds = [obj.find_all('td', class_='Table2__td') for obj in espn_tbodies]
espn_tds[0]
```
